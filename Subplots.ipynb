```python
from matplotlib import pyplot as plt
import numpy as np

x=np.arange(0.0,5.0,0.1)
y1=np.sin(2*np.pi*x)
y2=np.cos(2*np.pi*x)
plt.subplot(211)
plt.plot(x,y1,'r')
plt.subplot(212)
plt.plot(x,y2,'g')
plt.show()
```

```python
x=np.arange(0.0,5.0,0.1)
y1=np.sin(2*np.pi*x)
y2=np.cos(2*np.pi*x)

plt.figure(1)
plt.subplot(121)
plt.plot(x,y1,'o--r')
plt.grid('true')

plt.figure(1)
plt.subplot(122)
plt.plot(x,y2,'v:g')

plt.figure(2)
plt.subplot(211)
plt.plot(x,y1,'.-.y')

plt.figure(2)
plt.subplot(212)
plt.plot(x,y2,'^:b')
plt.grid('true')

plt.show()

```

```python
###Generate sine and cosine plot over a time scale of 0 to 5 with an interval of 0.5. 
###Plot them on two different subplots (12x) and apply xlabel, ylabel, Grids as well as the title.

from matplotlib import pyplot as pt
import numpy as np

x=np.sin(np.pi*0.5*t) ##multiplying by t to convert scalar values into array
y=np.cos(np.pi*0.5*t) ##multiplying by t to convert scalar values into array
t=np.arange(0.0,5.0,0.5)
pt.xlabel("time-scale")
pt.ylabel("angle")
pt.grid('true')
pt.title("Sin-Cos Plot")

pt.subplot(121)
pt.plot(x,t)
pt.xlabel("time-scale")
pt.ylabel("angle")
pt.grid('true')
pt.title("Sin-Cos Plot")
pt.figure(1)
pt.subplot(122)
pt.plot(y,t)
pt.xlabel("time-scale")
pt.ylabel("angle")
pt.grid('true')
pt.title("Sin-Cos Plot")


pt.show()
```

```python
##categorical x-y plot where y is subcategorized- will work for non-continuos
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
g = np.array([1,3,2,2])
plt.bar(x,y)
plt.bar(x,g,bottom=y)

plt.show()
```

```python
##categorical 
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y,explode=(0,0,0.1,0))

plt.show()
```
